```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('NBA_Players.csv')
```

# Limpeza do Dado
Aqui começo a fazer a limpeza do dado para que fique melhor a análise.

```python
dfT = df.drop(columns=['TEAM', ' COLLEGE', ' THM_THA', ' HT', ' WT', ' AGE', ' URL', ' THP'])

filtroSC = dfT[' NAME'] == 'Stephen Curry'

filtroStephen = dfT[filtroSC]
```

```python
pd.options.display.max_rows = None
pd.options.display.max_columns = None

dfT[' SALARY'] = dfT[' SALARY'].apply(lambda x: x.replace(',', ''))

dfT[' SALARY'] = dfT[' SALARY'].apply(lambda x: None if x=="Not signed" else x)

dfT[' SALARY'] = dfT[' SALARY'].astype(float)

dfT.sort_values(by=' SALARY', na_position='last', inplace=True)

positionPG = dfT[dfT[' POSITION'].str.contains('PG')]

positionPG.dropna()
```

```python
experience = positionPG.loc[positionPG[' EXPERIENCE'] >= 9]

mediaJogadores = positionPG.loc[positionPG[' EXPERIENCE'] >= 9]
```

```python
PPG_Stephen = filtroStephen[' PPG_LAST_SEASON'].mean()
PPG_Stephen = round(PPG_Stephen, 2)

PPGCARRER_Stephen = filtroStephen[' PPG_CAREER'].mean()
PPGCARRER_Stephen = round(PPGCARRER_Stephen, 2)

APG_Stephen = filtroStephen[' APG_LAST_SEASON'].mean()
APG_Stephen = round(APG_Stephen, 2)

APGCARRER_Stephen = filtroStephen[' APG_CAREER'].mean()
APGCARRER_Stephen = round(APGCARRER_Stephen, 2)

RPG_Stephen = filtroStephen[' RPG_LAST_SEASON'].mean()
RPG_Stephen = round(RPG_Stephen, 2)

RPGCARRER_Stephen = filtroStephen[' RGP_CAREER'].mean()
RPGCARRER_Stephen = round(RPGCARRER_Stephen, 2)

PER_Stephen = filtroStephen[' PER_LAST_SEASON'].mean()
PER_Stephen = round(PER_Stephen, 2)

GP_Stephen = filtroStephen[' GP'].mean()
GP_Stephen = round(GP_Stephen, 2)

MPG_Stephen = filtroStephen[' MPG'].mean()
MPG_Stephen = round(MPG_Stephen, 2)

FGP_Stephen = filtroStephen[' FGP'].mean()
FGP_Stephen = round(FGP_Stephen, 2)

FTP_Stephen = filtroStephen[' FTP'].mean()
FTP_Stephen = round(FTP_Stephen, 2)

APG_Stephen = filtroStephen[' APG'].mean()
APG_Stephen = round(APG_Stephen, 2)

BLKPG_Stephen = filtroStephen[' BLKPG'].mean()
BLKPG_Stephen = round(BLKPG_Stephen, 2)

STLPG_Stephen = filtroStephen[' STLPG'].mean()
STLPG_Stephen = round(STLPG_Stephen, 2)

TOPG_Stephen = filtroStephen[' TOPG'].mean()
TOPG_Jogadores = round(TOPG_Stephen, 2)

PPGJogo_Stephen = filtroStephen[' PPG'].mean()
PPGJogo_Stephen = round(PPGJogo_Stephen, 2)
```

<h4>Para melhor visualização e usabilidade resolvi atribuir os valores a variáveis com abreviações expecificas da tabela</h4>

<h4>Sumário</h4>
PPG = Points Per Game

APG = Assistance Per Game 

RPG = Rebounds Per Game

FGA = Field Goals Attempt

FTM = Free Throws Made

```python
PPG = [PPG_Jogadores, PPGCARRER_Jogadores]
PPGS = [PPG_Stephen, PPGCARRER_Stephen]

APG = [APG_Jogadores, APGCARRER_Jogadores]
APGS = [APG_Stephen, APGCARRER_Stephen]

RPG = [RPG_Jogadores, RPGCARRER_Jogadores]
RPGS = [RPG_Stephen, RPGCARRER_Stephen]

PER = [PER_Jogadores]
PERS = [PER_Stephen]

GP = [GP_Jogadores]
GPS = [GP_Stephen]

MPG = [MPG_Jogadores]
MPGS = [MPG_Stephen]

FGP = [FGP_Jogadores]
FGPS = [FGP_Stephen]

FTP = [FTP_Jogadores]
FTPS = [FTP_Stephen]

APG = [APG_Jogadores]
APGS = [APG_Stephen]

BLKG = [BLKPG_Jogadores]
BLKGS = [BLKPG_Stephen]

STLPG = [STLPG_Jogadores]
STLPGS = [STLPG_Stephen]

TOPG = [TOPG_Jogadores]
TOPGS = [TOPG_Stephen]

PPGJogo = [PPGJogo_Jogadores]
PPGSJogo = [PPGJogo_Stephen]

Estatisticas_Jogadores = [MPG[0], FGP[0], FTP[0], APG[0], BLKG[0], STLPG[0], TOPG[0], PPGJogo[0]]
Estatisticas_Stephen = [MPGS[0], FGPS[0], FTPS[0], APGS[0], BLKGS[0], STLPGS[0], TOPGS[0], PPGSJogo[0]]
```

```python
def autolabel(grupos):
     for i in grupos:
        h = i.get_height()
        ax.annotate('{}'.format(h),
                     xy = (i.get_x()+i.get_width()/2,h),
                     xytext = (0,3),
                     textcoords = 'offset points',
                     ha = 'center')
```

```python
largura = 0.35
tema = ['Temporada passada', 'Toda a carreira']
```

# Gráficos

A partir daqui vamos análisar os gráficos para saber como é o desempenho do Stephen de acordo com a média dos outros jogadores da mesma posição.

```python
x = np.arange(len(tema))

fig,ax = plt.subplots(figsize = (5,5))

mediaJogadoresPPG = ax.bar(x -largura/2,PPG,largura,label = 'Média dos jogadores')
stephenPPG = ax.bar(x +largura/2,PPGS,largura,label = 'Stephen Curry')

ax.set_title('Pontuação')
ax.legend()
ax.set_ylim([0, 30])
ax.set_xticks(x)
ax.set_xticklabels(tema)

autolabel(mediaJogadoresPPG)
autolabel(stephenPPG)

plt.show()
```

```python
fig,ax = plt.subplots(figsize = (5,5))
mediaJogadoresAPG = ax.bar(x -largura/2,APG,largura,label = 'Média dos jogadores')
stephenAPG = ax.bar(x +largura/2,APGS,largura,label = 'Stephen Curry')

ax.set_title('Assistências')
ax.legend()
ax.set_ylim([0, 10])
ax.set_xticks(x)
ax.set_xticklabels(tema)

autolabel(mediaJogadoresAPG)
autolabel(stephenAPG)

plt.show()
```

```python
fig,ax = plt.subplots(figsize = (5,6))
mediaJogadoresRPG = ax.bar(x -largura/2,RPG,largura,label = 'Média dos jogadores')
stephenRPG = ax.bar(x +largura/2,RPGS,largura,label = 'Stephen Curry')

ax.set_title('Rebotes')
ax.legend()
ax.set_ylim([0, 10])
ax.set_xticks(x)
ax.set_xticklabels(tema)

autolabel(mediaJogadoresAPG)
autolabel(stephenAPG)

plt.show()
```
